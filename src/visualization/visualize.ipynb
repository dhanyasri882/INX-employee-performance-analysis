## Visualization

### Objective
The objective of this is to present clear and intuitive visualizations
that summarize employee performance patterns and key influencing factors.
These visuals are designed for management-level understanding and decision-making.


### Purpose of Visualizations
The following charts present key insights in a simple and intuitive format
for management and HR decision-making.


### Load Processed Dataset
The processed employee performance dataset is loaded for visualization.This dataset was prepared during the data processing stage.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "../../data/processed/employee_performance_processed.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

### Employee Performance Rating Distribution
- This chart shows how employees are distributed across performance ratings.
- It helps management understand the overall performance landscape.


```python
plt.figure(figsize=(6,4))
sns.countplot(x='performance_rating', data=df)
plt.title('Employee Performance Rating Distribution')
plt.xlabel('Performance Rating')
plt.ylabel('Number of Employees')
plt.show()
```

### Insight
- Most employees fall within the average performance category.
- A smaller group of employees fall in lower and higher performance categories.
- This indicates an opportunity to support low performers while nurturing high performers.


### Average Performance by Department
- This chart compares the average performance rating across different departments.
- It helps identify departments that are performing well and those that may need support.


```python
department_performance = (
    df.groupby('department')['performance_rating']
      .mean()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))
department_performance.plot(kind='bar')
plt.title('Average Performance Rating by Department')
plt.xlabel('Department')
plt.ylabel('Average Performance Rating')
plt.xticks(rotation=45)
plt.show()
```

### Insight
- Performance levels vary across departments.
- High-performing departments can serve as benchmarks.
- Lower-performing departments may benefit from targeted training and support.


### Job Involvement vs Performance
This chart shows how job involvement varies across performance levels.
Higher involvement is expected to be associated with better performance.


```python
plt.figure(figsize=(6,4))
sns.boxplot(x='performance_rating', y='job_involvement', data=df)
plt.title('Job Involvement vs Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Job Involvement')
plt.show()
```

### Insight
Higher job involvement is generally associated with higher performance ratings.
This suggests employee engagement plays a key role in performance.


```python
plt.figure(figsize=(6,4))
sns.boxplot(x='performance_rating', y='work_life_balance', data=df)
plt.title('Work Life Balance vs Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Work Life Balance')
plt.show()
```

### Insight
- Employees with better work-life balance tend to show higher performance ratings.
- This highlights the importance of employee well-being initiatives.


```python
plt.figure(figsize=(6,4))
sns.boxplot(x='performance_rating', y='environment_satisfaction', data=df)
plt.title('Environment Satisfaction vs Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Environment Satisfaction')
plt.show()
```

### Insight
- Higher environment satisfaction is positively associated with employee performance.
- Improving workplace environment may lead to better productivity.


### Summary of Visual Insights
- Employee performance distribution shows most employees perform at average levels.
- Performance varies across departments.
- Job involvement, work-life balance, and environment satisfaction
  show strong relationships with performance.

These visuals support data-driven decision-making for improving employee performance.
