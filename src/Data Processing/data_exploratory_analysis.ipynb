## Exploratory Data Analysis

### Objective
The objective of this is to explore the processed employee dataset to
identify patterns, trends, and relationships that influence employee performance.
The insights derived from this analysis will support data-driven HR decisions.


### Load Processed Dataset
In this step, the cleaned and processed employee performance dataset created
in data processing is loaded. This dataset will be used for exploratory data analysis.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')

file_path = "../../data/processed/employee_performance_processed.csv"
df = pd.read_csv(file_path)
```

### Dataset overview

```python
df.head()
```

```python
df.describe()
```

#### Observation:

- The dataset contains numerical representations of all features.
- Performance Rating is a discrete target variable.

```python
df.columns
```

### Distribution of Employee Performance Rating

```python
plt.figure()
sns.countplot(x='performance_rating', data=df)
plt.title('Distribution of Employee Performance Ratings')
plt.xlabel('Performance Rating')
plt.ylabel('Employee Count')
plt.show()
```

 ### Insight: Performance Distribution
- The majority of employees fall in the average performance category.
- A smaller proportion of employees fall into high and low performance groups.
- This indicates an opportunity for performance improvement programs.


### Department-wise Performance Analysis
- employee performance is analyzed across different departments.
This helps identify departments with higher or lower average performance
and supports targeted improvement strategies.

```python
department_performance = df.groupby('department')['performance_rating'].mean().sort_values(ascending=False)
department_performance
```

```python
plt.figure()
department_performance.plot(kind='bar')
plt.title('Average Performance Rating by Department')
plt.xlabel('Department')
plt.ylabel('Average Performance Rating')
plt.show()
```

### Insight: Department-wise Performance
- The analysis shows clear variation in performance across departments.
- Some departments consistently achieve higher average ratings,while others show lower performance.

  Departments with lower average performance may require:
- Additional training programs
- Improved leadership support
- Better resource allocation


### Correlation Analysis

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title('Correlation Heatmap of Employee Attributes')
plt.show()
```

#### Insight
- Work-related attributes such as job involvement and environment satisfaction show strong correlation with performance.
- Compensation alone is not a strong predictor of performance.

### Factors Influencing Employee Performance
- key employee-related factors are analyzed against
  performance rating to understand their relationship with employee performance.

```python
key_features = ['job_involvement', 'environment_satisfaction', 'work_life_balance']
plt.figure(figsize=(6,4))
sns.boxplot(x='performance_rating', y='job_involvement', data=df)
plt.title('Job Involvement vs Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Job Involvement')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='performance_rating', y='work_life_balance', data=df)
plt.title('Work Life Balance vs Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Work Life Balance')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='performance_rating', y='environment_satisfaction', data=df)
plt.title('Environment Satisfaction vs Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Environment Satisfaction')
plt.show()
```

#### Insight
- Employees with higher job involvement generally show better performance.
- Better work-life balance is associated with higher performance ratings.
- Higher environment satisfaction correlates positively with employee performance.
- These factors indicate that employee engagement and workplace conditions
  play a significant role in driving performance.


### Business Interpretation
- Exploratory analysis suggests that employee engagement and workplace
conditions influence performance more than demographic factors.
- This supports the company’s focus on improvement rather than penalization.


### Summary of Exploratory Data Analysis
- Employee performance distribution was analyzed.
- Department-wise average performance differences were identified.
- Key factors such as job involvement, work-life balance,
  and environment satisfaction show a strong relationship with performance.
- These insights form the foundation for feature selection and predictive modeling.
